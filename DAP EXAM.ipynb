```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

# Load the dataset
df = pd.read_csv('2021_rws.csv', encoding='Windows-1252')

# Convert birth year to age
current_year = datetime.now().year
df['Age'] = current_year - df['What year were you born?  '].astype(int)

# Plot age distribution
plt.figure(figsize=(10, 6), facecolor='white')
plt.hist(df['Age'], bins=20, color='blue', alpha=0.7)
plt.title('Age Distribution of Respondents')
plt.xlabel('Age')
plt.ylabel('Number of Respondents')
plt.grid(True)
plt.show()
```

```python
# Plot gender distribution
plt.figure(figsize=(10, 6), facecolor='white')
df['What is your gender?  '].value_counts().plot(kind='bar', color='green', alpha=0.7)
plt.title('Gender Distribution of Respondents')
plt.xlabel('Gender')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
# Compare actual remote work to preferred remote work
plt.figure(figsize=(12, 6), facecolor='white')

# Actual remote work this year
actual_remote = df['Thinking about your current job, how much of your work time have you spent working remotely this year?  If you work a 5 day week, each day of remote working equals 20% of your time.  ']

# Preferred remote work this year
preferred_remote = df['How much of your work time would you have preferred to work remotely so far this year?  If you work a 5 day week, each day of remote working equals 20% of your time.  ']

# Plot
plt.hist([actual_remote, preferred_remote], bins=5, color=['red', 'blue'], alpha=0.7, label=['Actual Remote Work', 'Preferred Remote Work'])
plt.title('Comparison of Actual vs. Preferred Remote Work Percentages')
plt.xlabel('Percentage of Remote Work')
plt.ylabel('Number of Respondents')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot industry representation
plt.figure(figsize=(12, 8), facecolor='white')
df['Which of the following best describes your industry?  '].value_counts().plot(kind='barh', color='purple', alpha=0.7)
plt.title('Industry Representation in Survey')
plt.xlabel('Number of Respondents')
plt.ylabel('Industry')
plt.grid(True)
plt.show()
```

```python
# Analyze changes in remote work policy due to COVID-19
plt.figure(figsize=(10, 6), facecolor='white')
df['These questions are about the remote working arrangements in your organisation.  Has your employer changed or updated their policy on remote work since COVID-19?   This means since February/March 2020  '].value_counts().plot(kind='bar', color='orange', alpha=0.7)
plt.title('Changes in Employer Remote Work Policy Since COVID-19')
plt.xlabel('Response')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
